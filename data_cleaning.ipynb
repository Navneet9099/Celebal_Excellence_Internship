```python
import pandas as pd

df = pd.read_csv('archive/backpacks.csv')

print("Dataset loaded successfully!")
print(df.head())
```

```python
# How many rows and columns?
print("Shape:", df.shape)

# What are the column names?
print("\nColumns:", df.columns.tolist())

# What type of data is in each column?
print("\nData Types:")
print(df.dtypes)

# Full overview
print("\nFull Info:")
df.info()

# Statistical summary (min, max, average, etc.)
print("\nStatistics:")
df.describe()
```

```python
# First, check HOW MANY missing values are in each column
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Fill text/string columns with a placeholder word
df['what_customers_said'] = df['what_customers_said'].fillna('No Review')
df['seller_name'] = df['seller_name'].fillna('Unknown Seller')
df['seller_information'] = df['seller_information'].fillna('No Info')
df['variations'] = df['variations'].fillna('No Variation')

# Verify - all should show 0 now
print("After handling missing values:")
print(df.isnull().sum())
```

```python
print(df.columns.tolist())
```

```python
# Filter 1: Show products with rating greater than 4
high_rated = df[df['rating'] > 4]
print("High rated products (rating > 4):")
print(high_rated.shape)  # how many rows matched
print(high_rated[['title', 'rating', 'final_price']].head())
```

```python
# Filter 2: Show products that have a discount
discounted = df[df['discount'] > 0]
print("\nDiscounted products:")
print(discounted.shape)
print(discounted[['title', 'discount', 'final_price']].head())
```

```python
# Select only important columns
important_cols = df[['product_id', 'title', 'rating', 'initial_price', 'final_price', 'discount', 'category']]

print("Selected columns only:")
print(important_cols.head(10))
```

```python
# Show title, price, rating of only high-rated products
result = df[df['rating'] > 4][['title', 'rating', 'final_price']]
print(result.head(10))
```

```python
# Count how many duplicate rows exist
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
df = pd.read_csv('archive/Combined_dataset.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Check if 'quantity' column is there
if 'quantity' in df.columns:
    print("✅ quantity column EXISTS!")
else:
    print("❌ No quantity column")
    
print(df.columns.tolist())
```

```python
import numpy as np

# Create quantity column (simulated - 1 to 10 units)
np.random.seed(42)
df['quantity'] = np.random.randint(1, 11, size=len(df))

# Create total_amount = final_price × quantity
df['total_amount'] = df['final_price'] * df['quantity']

# Verify
print("✅ New columns created!")
print(df[['title', 'final_price', 'quantity', 'total_amount']].head(10))
```

### Note on total_amount column

This dataset is a product listing dataset (similar to Amazon/Flipkart catalog).
It does not contain a quantity column as products are listed individually.

To fulfil the assignment requirement of:
total_amount = price × quantity

We simulated a quantity column with random values (1–10) representing
hypothetical units sold per product.

Formula used: total_amount = final_price × quantity

```python
print(df['final_price'].dtype)
print(df['final_price'].head())
```

```python
# Remove ₹ symbol, commas, quotes and convert to float
df['final_price_clean'] = df['final_price'].astype(str)
df['final_price_clean'] = df['final_price_clean'].str.replace('₹', '', regex=False)
df['final_price_clean'] = df['final_price_clean'].str.replace(',', '', regex=False)
df['final_price_clean'] = df['final_price_clean'].str.replace('"', '', regex=False)
df['final_price_clean'] = df['final_price_clean'].str.strip()
df['final_price_clean'] = pd.to_numeric(df['final_price_clean'], errors='coerce')

print("Cleaned prices:")
print(df['final_price_clean'].head(10))
print("Data type:", df['final_price_clean'].dtype)
```

```python
df['total_amount'] = df['final_price_clean'] * df['quantity']

print("\n✅ total_amount created correctly!")
print(df[['title', 'final_price_clean', 'quantity', 'total_amount']].head(10))
```

```python
# Final check - see what your cleaned dataframe looks like
print("Final Shape:", df.shape)
print("\nColumns in cleaned dataset:", df.columns.tolist())
print("\nAny missing values left?")
print(df.isnull().sum())
```

```python
# Save the cleaned dataset as a new CSV file
df.to_csv('cleaned_shopping_dataset.csv', index=False)

print("✅ File saved successfully!")
print("Filename: cleaned_shopping_dataset.csv")
print("Total rows saved:", len(df))
print("Total columns saved:", len(df.columns))
```
